```python
page1=requests.get("https://www.imdb.com/chart/top")
page1
```

```python
soup1=BeautifulSoup(page1.content,'html.parser')
soup1
```

```python
name=soup1.find_all('td',class_="titleColumn")
name
```

```python
movies=[]
for movie in name:
    movie=movie.get_text().replace('\n',"")
    movie=movie.strip(" ")
    movies.append(movie)
movies
```

```python
rate=soup1.find_all("td",class_="ratingColumn imdbRating")
rate
```

```python
ratings=[]
for rating in rate:
    rating=rating.get_text().replace('\n','')
    ratings.append(rating)
ratings
```

```python
years=soup1.find_all("span",class_="secondaryInfo")
years
```

```python
year_of_releases=[]
for year_of_release in years:
    year_of_release=year_of_release.get_text().replace('/n','')
    year_of_releases.append(year_of_release)
year_of_releases
```

```python
import pandas as pd
```

```python
data= pd.DataFrame()
data["movie name"]=movies
data['Ratings']=ratings
data['year_of_release']=year_of_releases
data.head(100)
```

```python
page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
titles=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
titles
```

```python
location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
RS=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    RS.append(i.text)
RS
```

```python
images=[]
for i in soup.find_all("img",class_="no-img"):
    images.append(i['data-src'])
images
```

```python
rattings=[]
for i  in soup.find_all("div",class_="restnt-rating rating-4"):
    rattings.append(i.text)
rattings
```

```python
df=pd.DataFrame({"titles":titles,"location":location,"Price":RS,"images_url":images,"ratings":rattings})
df
```
